```python
import json; import os; import pandas as pd
```

#### Opens Json files and converts to list of dictionaries

```python
def jsonContents(filePath):
    fileData = []; all_possible_keys = set(); main_data = {}
    contents = filePath.replace('\\','/')
    with open(contents,encoding="utf8") as f:
        data = json.load(f) 
        fileData.append(data)
        for d in data:
           for k in d.keys():
               all_possible_keys.add(k)
    
    for stuff in data:
        for field in list(all_possible_keys):
            if field not in main_data:
                main_data[field] = []
            if field in stuff:
                main_data[field].append(stuff[field])
            else:
                main_data[field].append(None)
    dataframe = pd.DataFrame.from_dict(main_data)
    return dataframe
```

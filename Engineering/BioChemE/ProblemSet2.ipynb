```python
### Problem 2 ###
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Energy constant in j/mol*K
R = 8.314

# x-axis: T converted to Kelvin [1/T]
T = 1/np.array([40.4+273,36.8+273,33.1+273,30.0+273,25.1+273]).reshape(-1,1)

# y-axis: Rate constant "k" value [ln(k)]
Rate_Constant = np.log(np.array([0.014,0.0112,0.0074,0.0051,0.0036]))

### Fitting data 
model = LinearRegression().fit(T,Rate_Constant)
Rate_model = []

for k_values in T:
    modelEqnSoln = model.coef_*k_values + model.intercept_
    Rate_model.append(modelEqnSoln)
np.array(Rate_model)
###

### Plotting and regressing using "Arrhenius equation" 
plt.plot(T,Rate_Constant,'o', label = "Rate Data")
plt.plot(T,Rate_model, label = "Rate model")
plt.title('Arrhenius plot')
plt.ylabel('ln(k)')
plt.xlabel('1/T')
plt.legend()
plt.show()
print('The slope is: ',round(model.coef_[0],2),"Therefore the activation energy 'Ea' is:",round(model.coef_[0]*R/1000,2),"kj/mol*K")
###
```

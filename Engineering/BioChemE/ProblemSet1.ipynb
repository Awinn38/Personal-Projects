```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# x-axis
s0 = np.array([0.1,0.04,0.02,0.01,0.005,0.004]).reshape(-1,1)
# y-axis
Vnodiff = np.array([0.05,0.025,0.017,0.0125,0.011,0.0108])
# Fitting data 
model = LinearRegression().fit(s0,Vnodiff)
Vmodel = []

for substrateConc in s0:
    modelEqnSoln = model.coef_*substrateConc + model.intercept_
    Vmodel.append(modelEqnSoln)
np.array(Vmodel)

# Plotting and regressing using "Lineweaver-Burk plot" based on assumptions
plt.plot(s0,Vnodiff,'o', label = "Velocity Data")
plt.plot(s0,Vmodel, label = "Velocity model")
plt.legend()
plt.show()
print('The x-intercept is: ',model.intercept_)
print('The slope is: ',model.coef_)
print('The value of Vm is: ',(1/model.intercept_),'mg/L')
print('The value of Km is: ',(model.coef_*1/model.intercept_),'mg/L')
```
